```python
import cv2
import os
import matplotlib.pyplot as plt
import pandas as pd
```

Based on https://www.pyimagesearch.com/2015/09/07/blur-detection-with-opencv/

```python
def calculate_blur(img):
    return cv2.Laplacian(img, cv2.CV_64F).var()
```

```python
lst = os.listdir('data/scenes/cars_3')
dct = {}
for img_name in lst:
    if not os.path.isdir(img_name):
        img = cv2.imread(os.path.join('data/scenes/cars_3', img_name))
        if type(img) != type(None):
            dct[img_name] = calculate_blur(img)
df = pd.Series(dct)
```

```python
display(df.argmax(), df.values.max())
```

```python
display(df.argmin(), df.values.min())
```
