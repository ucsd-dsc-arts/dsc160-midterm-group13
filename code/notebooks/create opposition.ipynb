```python
import cv2
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def t(img):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    img[:, :, 2] = img[:, :, 2]/1.5
    img[:, :, 1] = img[:, :, 1]/1.3
    img = cv2.cvtColor(img, cv2.COLOR_HSV2BGR)
    return img
```

```python
def resize(img):
    return cv2.resize(img, (128,256))
```

```python
clean = pd.read_csv('logistic_regression_trained_on_outside.csv')
joys = clean[clean.predict == 'joy']
import imageio
images = []
path1 = list(joys.paths_1)
path2 = list(joys.paths_2)
path3 = list(joys.paths_3)
images = []
o = []
for i in range(len(path1)):
    images.append(resize(imageio.imread(path1[i])))
    images.append(resize(imageio.imread(path2[i])))
    images.append(resize(imageio.imread(path3[i])))
    o.append(resize(t(imageio.imread(path1[i]))))
    o.append(resize((imageio.imread(path2[i]))))
    o.append(resize((imageio.imread(path3[i]))))
kargs = { 'duration': 0.5}
imageio.mimsave('joy_small.gif', images, 'GIF', **kargs)
imageio.mimsave('not_joy_small.gif', o, 'GIF', **kargs)
```
