This function resize one single image(its path) to the proportion of orignally one. Some image from scratch maybe none, it will ignore it

```python
def resize_one(img, factor):
    img = cv2.imread(img)
    if type(img) == 'NoneType:
        return
    shape = img.shape
    new_shape = (int(shape[1]/factor), int(shape[0]/factor))
    img = cv2.resize(img, new_shape)
    return img
```

This function takes one movie dir and the expected resize factor (proportion of orginal size) and store the new images in new dirctory. Please do not submit resize images to github unless everyone agree.

```python
def resize_one_movie(movie_dir, factor):
    new_dir = os.path.join(movie_dir, 'resize{x}'.format(x=factor))
    if os.path.isdir(new_dir) == False:
        os.mkdir(new_dir)
    for file in os.listdir(movie_dir):
        if os.path.isdir(file) == False:
            img_path = os.path.join(new_dir, file)
            new_img = resize_one(os.path.join(movie_dir, file), factor)
            if type(new_img) != None:
                cv2.imwrite(img_path, new_img) 
```

```python
resize_one_movie(data, 5)
```
