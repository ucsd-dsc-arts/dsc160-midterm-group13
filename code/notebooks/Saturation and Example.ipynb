```python
import cv2
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def mean_saturation(img):
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV) #convert it to hsv
    return np.mean(hsv[:,:,1])
```

```python
lst = os.listdir('data/scenes/big_hero_6')
dct = {}
for img_name in lst:
    if not os.path.isdir(img_name):
        img = cv2.imread(os.path.join('data/scenes/big_hero_6', img_name))
        if type(img) != type(None):
            dct[img_name] = mean_saturation(img)
df = pd.Series(dct)
display(df.argmin(), df.values.min())
```

```python
display(df.argmax(), df.values.max())
```
